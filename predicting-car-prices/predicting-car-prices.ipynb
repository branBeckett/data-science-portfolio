# Predicting Car Prices

In this projects, we'll use several machine learning models to predict a car's market price.

The dataset we'll be working with contains information about technical aspects of various vehicles like motor displacement, weight, miles per gallon, acceleration, etc. More information on the data can be found [here](https://archive.ics.uci.edu/ml/datasets/automobile).

## Exploring the Data

Let's start off by getting more familiar with the dataset.

```python
import pandas as pd
import numpy as np

pd.options.display.max_columns = 100
```

```python
# There are no column headers in our dataset, so we will add our own based off of dataset's documentation
column_names = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 
        'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 
        'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-rate', 'horsepower',
        'peak-rpm', 'city-mpg', 'highway-mpg', 'price']

cars = pd.read_csv('imports-85.data', names=column_names)
```

```python
cars.head()
```

```python
# Selecting only the numeric columns that can be used as features and selecting our target column for predictions
numeric_columns = ['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-size',
                   'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg',
                   'price']
numeric_cars = cars[numeric_columns]
```

## Cleaning the Data

Next, we'll need to remove any missing values since we can't have any in our dataset for predictive modeling.

```python
# Replace the "?" in the normalized-losses column
numeric_cars = numeric_cars.replace('?', np.nan)
numeric_cars.head()
```

The "?" that we just converted to null values were originally cast to the pandas object as a string, so these columns need to be converted back into a numeric type.

```python
numeric_cars = numeric_cars.astype('float')
```

```python
numeric_cars.isnull().sum()
```

We can't have any missing values in our dataset, and because `price` is going to be the column we are predicting, we will remove these rows.

```python
# Removing rows with missing values
numeric_cars = numeric_cars.dropna(subset=['price'])
numeric_cars.isnull().sum()
```

We'll replace the missing values in the other columns with that column's mean value.

```python
numeric_cars = numeric_cars.fillna(numeric_cars.mean())
numeric_cars.isnull().sum()
```

Our last step in cleaning the data before training our model is to normalize all of our columns except `price` so that their values fall between 0 and 1.

```python
# Normalizing the data
price_save = numeric_cars['price']
numeric_cars = (numeric_cars - numeric_cars.min()) / (numeric_cars.max() - numeric_cars.min())
numeric_cars['price'] = price_save
```

## Univariate Model

We'll start out with univariate k-nearest neighbors models. Beginning with simpler models can help us better structure our workflow and understand our features before moving on to more complex models.

```python
# Univariate k-nearest neighbors model with default k value

from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

def knn_train_test(train_col, target_col, df_obj):
    knn = KNeighborsRegressor()
    np.random.seed(1)
    
    # Randomize the rows in our dataframe
    index_shuffled = np.random.permutation(df_obj.index)
    random_df = df_obj.reindex(index_shuffled)
    
    # Split number of rows in half and round
    train_row = int(len(random_df) / 2)
    
    # Select training and test sets
    train_df = random_df.iloc[0:train_row]
    test_df = random_df.iloc[train_row:]
    
    # Fit the KNN model with default k value
    knn.fit(train_df[[train_col]], train_df[target_col])
    
    # Making predictions
    predicted = knn.predict(test_df[[train_col]])
    
    # Calculate RMSE
    mse = mean_squared_error(test_df[target_col], predicted)
    rmse = np.sqrt(mse)
    return rmse

rmse_data = {}

# For each column except 'price', train the model and return the RMSE values
train_cols = numeric_cars.columns.drop('price')

for col in train_cols:
    rmse_val = knn_train_test(col, 'price', numeric_cars)
    rmse_data[col] = rmse_val
    
# Create a Series object from the dictionary to more easily view the results
rmse_data_series = pd.Series(rmse_data)
rmse_data_series.sort_values()
```

```python
# Univariate k-nearest neighbors model with various k values

def knn_train_test(train_col, target_col, df_obj):
    np.random.seed(1)
    
    # Randomize the rows in our dataframe
    index_shuffled = np.random.permutation(df_obj.index)
    random_df = df_obj.reindex(index_shuffled)
    
    # Split number of rows in half and round
    train_row = int(len(random_df) / 2)
    
    # Select training and test sets
    train_df = random_df.iloc[0:train_row]
    test_df = random_df.iloc[train_row:]
    
    # Fit the KNN model with several k values
    k_values = [1, 3, 5, 7, 9]
    k_rmse = {}
    
    for k in k_values:
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(train_df[[train_col]], train_df[target_col])
    
        # Making predictions
        predicted = knn.predict(test_df[[train_col]])
    
        # Calculate RMSE
        mse = mean_squared_error(test_df[target_col], predicted)
        rmse = np.sqrt(mse)
        
        k_rmse[k] = rmse
    
    return k_rmse

k_rmse_data = {}

# For each column except 'price', train the model and return the RMSE values
train_cols = numeric_cars.columns.drop('price')

for col in train_cols:
    rmse_val = knn_train_test(col, 'price', numeric_cars)
    k_rmse_data[col] = rmse_val
    
k_rmse_data
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')

for k, v in k_rmse_data.items():
    x = list(v.keys())
    y = list(v.values())
    
    plt.plot(x, y)
    plt.xlabel('k value')
    plt.ylabel('RMSE')
```

## Multivariate Model

Now we'll modify the `knn_train_test()` function we wrote so that it works with multiple columns.

```python
# Calculate the average RMSE across different values of k for each feature

feature_avg_rmse = {}

for k, v in k_rmse_data.items():
    avg_rmse = np.mean(list(v.values()))
    feature_avg_rmse[k] = avg_rmse

series_avg_rmse = pd.Series(feature_avg_rmse)
sorted_series_avg_rmse = series_avg_rmse.sort_values()

print(sorted_series_avg_rmse)

sorted_features = sorted_series_avg_rmse.index
```

```python
# Multivariate k-nearest neighbors model

def knn_train_test(train_col, target_col, df_obj):
    np.random.seed(1)

    # Randomize the rows in our dataframe
    index_shuffled = np.random.permutation(df_obj.index)
    random_df = df_obj.reindex(index_shuffled)
    
    # Split number of rows in half and round
    train_row = int(len(random_df) / 2)
    
    # Select training and test sets
    train_df = random_df.iloc[0:train_row]
    test_df = random_df.iloc[train_row:]
    
    # Fit the KNN model with several k values
    k_values = [5]
    k_rmse = {}
    
    for k in k_values:
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(train_df[train_col], train_df[target_col])
    
        # Making predictions
        predicted = knn.predict(test_df[train_col])
    
        # Calculate RMSE
        mse = mean_squared_error(test_df[target_col], predicted)
        rmse = np.sqrt(mse)
        
        k_rmse[k] = rmse
    
    return k_rmse

k_rmse_data = {}
    
for num_features in range(2,7):
    k_rmse_data['{} best features'.format(num_features)] = knn_train_test(
        sorted_features[:num_features],
        'price',
        numeric_cars
    )

k_rmse_data
```

## Hyperparameter Tuning

Now let's optimize the best performing models that we've found. We'll vary the hyperparameter value from 1 to 25 and plot the resulting RMSE values.

```python
# Multivariate k-nearest neighbors model with various k values

def knn_train_test(train_col, target_col, df_obj):
    np.random.seed(1)

    # Randomize the rows in our dataframe
    index_shuffled = np.random.permutation(df_obj.index)
    random_df = df_obj.reindex(index_shuffled)
    
    # Split number of rows in half and round
    train_row = int(len(random_df) / 2)
    
    # Select training and test sets
    train_df = random_df.iloc[0:train_row]
    test_df = random_df.iloc[train_row:]
    
    # Fit the KNN model with several k values
    k_values = [i for i in range(1, 25)]
    k_rmse = {}
    
    for k in k_values:
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(train_df[train_col], train_df[target_col])
    
        # Making predictions
        predicted = knn.predict(test_df[train_col])
    
        # Calculate RMSE
        mse = mean_squared_error(test_df[target_col], predicted)
        rmse = np.sqrt(mse)
        
        k_rmse[k] = rmse
    
    return k_rmse

k_rmse_data = {}
    
for num_features in range(2,6):
    k_rmse_data['{} best features'.format(num_features)] = knn_train_test(
        sorted_features[:num_features],
        'price',
        numeric_cars
    )

k_rmse_data
```

```python
for k,v in k_rmse_data.items():
    x = list(v.keys())
    y = list(v.values())  
    plt.plot(x,y, label="{}".format(k))
    
plt.xlabel('k value')
plt.ylabel('RMSE')
plt.legend()
```
